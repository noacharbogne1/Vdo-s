```python
import pandas as pd 
import os
import requests
import json
import time
import math
```

```python

url = "https://attrap.fr/api/v1/search?s={}&page={}&start_date=2025-09-05&end_date=2025-09-10&sort=relevance"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36"
}
searchs = ['aéronefs', 'drones']

page = 1
for search in searchs:
    response = requests.get(url.format(search, page), headers=headers)
    data = response.json()
    # Process the data as needed
    total_pages = math.ceil(data['total'] / 10)
    with open(f"data/input/data_attrap_{search}_page_{page}.json", "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
    print(f"Total pages for '{search}': {total_pages}")
    for page in range(2, total_pages + 1):
        print(f"Fetching page {page} for '{search}'")
        response = requests.get(url.format(search, page), headers=headers)
        data = response.json()
        with open(f"data/input/data_attrap_{search}_page_{page}.json", "w", encoding="utf-8") as f:
            json.dump(data, f, ensure_ascii=False, indent=4)
        time.sleep(1)  # Be polite and avoid hitting the server too hard
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"
}

for f in os.listdir('./data/input'):
    print(f)
    if f.endswith('.json'):
        d = json.load(open(os.path.join('./data/input', f)))
        for el in d['elements']:
            url = el.get('url')
            pref = el.get('administration')
            id = el.get('id')
            filename = f"{pref}_{id}.pdf"
            try:
                r = requests.get(url, headers=headers, timeout=15)
                r.raise_for_status()
                with open(os.path.join('./data/arretes', filename), 'wb') as outfile:
                    outfile.write(r.content)
                print(f"Downloaded {filename}")
            except Exception as e:
                print(f"Error downloading {url}: {e}")
            time.sleep(1)
```
